```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define two vectors
v1 = np.array([0.267, 0.534, 0.802])
v2 = np.array([0.802, 0.267, 0.534])

# Compute Euclidean distance
distance = np.linalg.norm(v1 - v2)

# Create a new figure with a specific size (width, height)
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='b', linestyle="dashed", arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='r', linestyle="dashed", arrow_length_ratio=0.1)

# Draw a line between the vectors
ax.plot([v1[0], v2[0]], [v1[1], v2[1]], [v1[2], v2[2]], 'g', linewidth=3)

# Add labels for the vectors
ax.text(v1[0], v1[1], v1[2], 'v1', fontsize=12, ha='right')
ax.text(v2[0], v2[1], v2[2], 'v2', fontsize=12, ha='right')

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])

# Add a title
plt.title(f'Euclidean Distance: {distance:.2f}')

# Show the plot
plt.savefig("distance_euclidean_raw.svg", format='svg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define two vectors
v1 = np.array([1, 2, 3])
v2 = np.array([3, 1, 2])

# Compute Manhattan distance
distance = np.sum(np.abs(v1 - v2))

# Create a new figure with a specific size (width, height)
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='b', linestyle="dashed", arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='r', linestyle="dashed", arrow_length_ratio=0.1)

# Draw lines to represent the Manhattan distance
ax.plot([v1[0], v2[0]], [v1[1], v1[1]], [v1[2], v1[2]], 'g', linewidth=3)
ax.plot([v2[0], v2[0]], [v1[1], v2[1]], [v1[2], v1[2]], 'g', linewidth=3)
ax.plot([v2[0], v2[0]], [v2[1], v2[1]], [v1[2], v2[2]], 'g', linewidth=3)

# Add labels for the vectors
ax.text(v1[0], v1[1], v1[2], 'v1', fontsize=12, ha='right')
ax.text(v2[0], v2[1], v2[2], 'v2', fontsize=12, ha='right')

ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

# Add a title
plt.title(f'Manhattan Distance: {distance:.2f}')

# Show the plot
plt.savefig("distance_manhattan_raw.svg", format='svg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define two vectors
v1 = np.array([1, 2, 3])
v2 = np.array([3, 1, 2])

# Compute dot product
dot_product = np.dot(v1, v2)

# Create a new figure with a specific size (width, height)
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='b', linestyle="dashed", arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='r', linestyle="dashed", arrow_length_ratio=0.1)

# Calculate the projection of v1 onto v2
proj = (np.dot(v1, v2) / np.linalg.norm(v2)**2) * v2

# Plot the projection
ax.quiver(0, 0, 0, proj[0], proj[1], proj[2], color='g', linewidth=2, arrow_length_ratio=0.1)

# Plot a line connecting the projection to v1
ax.plot([v1[0], proj[0]], [v1[1], proj[1]], [v1[2], proj[2]], 'g--')


# Add labels for the vectors
ax.text(v1[0], v1[1], v1[2], 'v1', fontsize=12, ha='right')
ax.text(v2[0], v2[1], v2[2], 'v2', fontsize=12, ha='right')
ax.text(proj[0], proj[1], proj[2], 'proj', fontsize=12, ha='right')

ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])

# Add a title
plt.title(f'Dot Product: -{dot_product:.2f}')

# Show the plot
plt.savefig("distance_negativeinner_raw.svg", format='svg')
plt.show()
```

```python
import math

def normalize_vector(vector):
    x, y, z = vector
    magnitude = math.sqrt(x**2 + y**2 + z**2)
    return (x/magnitude, y/magnitude, z/magnitude)

vector = (3, 1, 2)
normalized_vector = normalize_vector(vector)
print(normalized_vector)
```
